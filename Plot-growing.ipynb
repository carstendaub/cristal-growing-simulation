```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/carsten.daub/Downloads/crystal_path.csv')

# Plot the path
plt.figure(figsize=(8, 8))
plt.plot(df['x'], df['y'], marker='o', linestyle='-')
plt.title('Crystal Growth Path')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```
